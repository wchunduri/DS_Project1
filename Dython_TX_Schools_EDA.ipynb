```python
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/nvamsimohan/DallasDSA/refs/heads/main/2023%20Texas%20High%20Schools%20Scores.csv')
data
```

```python
# Code snippet 3
# Displaying dataset columns
data.columns
```

```python
# Code snippet 4
# Qualitative (ignoring noise variables) and quantitative data preparation (ignoring variables with more than 20% missing data)
# Dropping 'School', 'School URL', 'District URL',  'City URL', 'Phone',  'Home Prices Ranking', 'Community Health Ranking' variables
# if you put columns = then dont need to use axes =1
predictors = data.drop(columns=['School', 'School URL', 'District URL',  'City URL', 'Phone',  'Home Prices Ranking', 'Community Health Ranking'])


# Display predictors columns
predictors.columns
```

```python
# code snippet 5
#setting zip datatype as string
predictors['Zip'] = predictors['Zip'].astype(str)
```

```python
# Code Snippet 6
#importing relavent statistical tests from Dython
from dython.nominal import associations
```

```python
#code snippet 7
#Running the associations function on the
assoc = associations(predictors)
```

```python
# code snippet 8
# storing the above associations heatmap as a matrix
assoc_matrix = assoc['corr']
```

```python
# Code snippet 9
# setting the target variable
target ='Standard Score'
```

```python
# Code snippet 10
#  Listing the most associated/impacting variables on the taarget variables
assoc_matrix[target].abs().sort_values(ascending=False)
```

```python
# Code snippet 11
# Setting the target variable (for unit testing)
target = 'Percent Free/Disc Lunch'
# Listing the most associated/impacting variables on the target variable
assoc_matrix[target].abs().sort_values(ascending=False)
```
