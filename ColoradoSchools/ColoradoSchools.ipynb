```python
# What is the spread/range of the student population across the schools in the dataset?
student_population_range = predictors['Number Students'].max() - predictors['Number Students'].min()
print(f"The spread/range of the student population is: {student_population_range}")
```

```python
#2. Are there any school districts where the student-to-teacher ratio exceeds 25, indicating a high workload for teachers?

high_ratio_districts = predictors[predictors['Student/Teacher Ratio'] > 25]['District'].unique()
if len(high_ratio_districts) > 0:
    print("School districts where the student-to-teacher ratio exceeds 25:")
    for district in high_ratio_districts:
        print(district)
else:
    print("No school districts found where the student-to-teacher ratio exceeds 25.")
```

```python
#3. What is the median percentage of students receiving free or discounted lunches in the schools?
predictors['Percent Free/Disc Lunch'] = predictors['Percent Free/Disc Lunch'].str.replace('%', '').astype(float)
median_free_lunch = predictors['Percent Free/Disc Lunch'].median()
print(f"The median percentage of students receiving free or discounted lunches is: {median_free_lunch}%")
```

```python
#4. What are the extreme cases in terms of student population size in the schools?
min_students = predictors['Number Students'].min()
max_students = predictors['Number Students'].max()

schools_min_students = data[data['Number Students'] == min_students]['School'].tolist()
schools_max_students = data[data['Number Students'] == max_students]['School'].tolist()

print(f"Schools with the minimum student population ({min_students} students): {', '.join(schools_min_students)}")
print(f"Schools with the maximum student population ({max_students} students): {', '.join(schools_max_students)}")
```

```python
#5. What is the average student-to-teacher ratio across the school districts?
average_student_teacher_ratio = predictors['Student/Teacher Ratio'].mean()
print(f"The average student-to-teacher ratio across school districts is: {average_student_teacher_ratio:.2f}")
```

```python
#6. What percentage of schools are Title I, Charter, Magnet, or Virtual schools?
total_schools = len(predictors)

categories = ['Is Title I', 'Is Charter', 'Is Magnet', 'Is Virtual']

for category in categories:
    # Count occurrences where the string value is exactly 'Yes'
    count_yes = predictors[category].eq('Yes').sum()
    percentage = (count_yes / total_schools) * 100
    print(f"Percentage of schools that are {category}: {percentage:.2f}%")
```

```python
#7.How many school cities are tracked in the dataset?
num_unique_cities = predictors['City'].nunique()
print(f"There are {num_unique_cities} unique school cities tracked in the dataset.")
```

```python
lowest_full_time_teachers = predictors['Number Full-time Teachers'].min()
print(f"The lowest number of full-time teachers seen in any school district is: {lowest_full_time_teachers}")

schools_with_lowest_teachers = data[data['Number Full-time Teachers'] == lowest_full_time_teachers][['School', 'District', 'Student/Teacher Ratio']]
print("\nSchool(s) and District(s) with the lowest number of full-time teachers:")
print(schools_with_lowest_teachers.to_string(index=False))
```

```python
#9. What is the maximum student-to-teacher ratio observed in any school? name the school and school district
max_student_teacher_ratio = predictors['Student/Teacher Ratio'].max()
print(f"The maximum student-to-teacher ratio observed in any school is: {max_student_teacher_ratio}")

schools_with_max_ratio = data[data['Student/Teacher Ratio'] == max_student_teacher_ratio][['School', 'District']]
print("\nSchool(s) and District(s) with the highest student-to-teacher ratio:")
print(schools_with_max_ratio.to_string(index=False))
```

```python
#10. What is the range of ‘Percent White’ across the schools? Name the school and school district.

# Convert 'Percent White' to numeric, handling potential errors if any exist
predictors['Percent White'] = predictors['Percent White'].str.replace('%', '', regex=False).astype(float)

min_percent_white = predictors['Percent White'].min()
max_percent_white = predictors['Percent White'].max()
range_percent_white = max_percent_white - min_percent_white

print(f"The range of 'Percent White' across the schools is: {range_percent_white:.2f}%")

schools_min_percent_white = data[data['Percent White'].str.replace('%', '', regex=False).astype(float) == min_percent_white][['School', 'District']]
schools_max_percent_white = data[data['Percent White'].str.replace('%', '', regex=False).astype(float) == max_percent_white][['School', 'District']]

print(f"\nSchool(s) and District(s) with the lowest 'Percent White' ({min_percent_white}%):")
print(schools_min_percent_white.to_string(index=False))

print(f"\nSchool(s) and District(s) with the highest 'Percent White' ({max_percent_white}%):")
print(schools_max_percent_white.to_string(index=False))
```

```python
# 9. Calculate the correlation between 'Percent White' and 'Average Standard Score (2024-25)'
correlation_white_score = predictors['Percent White'].corr(predictors['Average Standard Score (2024-25)'])
print(f"The correlation between 'Percent White' and 'Average Standard Score' is: {correlation_white_score:.2f}")
```

```python
# Calculate the correlation between 'Student/Teacher Ratio' and 'Average Standard Score'
correlation_ratio_score = predictors['Student/Teacher Ratio'].corr(predictors['Average Standard Score (2024-25)'])
print(f"The correlation between 'Student/Teacher Ratio' and 'Average Standard Score' is: {correlation_ratio_score:.2f}")
```

```python
#10. What is the maximum student-to-teacher ratio observed in any school?
min_avg_score = predictors['Average Standard Score (2024-25)'].min()
max_avg_score = predictors['Average Standard Score (2024-25)'].max()

schools_min_score = data[data['Average Standard Score (2024-25)'] == min_avg_score][['School', 'District', 'Average Standard Score (2024-25)']]
schools_max_score = data[data['Average Standard Score (2024-25)'] == max_avg_score][['School', 'District', 'Average Standard Score (2024-25)']]

print("\nSchool(s) with the lowest 'Average Standard Score':")
display(schools_min_score)

print("\nSchool(s) with the highest 'Average Standard Score':")
display(schools_max_score)
```
