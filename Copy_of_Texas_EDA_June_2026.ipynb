```python
# Code snippet 1
# Setting the dataset URL
url = 'https://raw.githubusercontent.com/nvamsimohan/DallasDSA/refs/heads/main/2023%20Texas%20High%20Schools%20Scores.csv'

# Display the url
url
```

```python
# Code snippet 2
# Importing pandas for data processing
import pandas
```

```python
# Code snippet 3
# Importing the dataset from URL using pandas
data = pandas.read_csv(url)

# Display the data
data
```

```python
# Code snippet 4
# Displaying dataset columns
data.columns
```

```python
# Code snippet 5
# Displaying descriptive stats
data.describe()
```

```python
# Code snippet 6
# Importing missingno package - to display the missing data proportion
import missingno as msno
```

```python
# Code snippet 7
# Displaying the missing data proportion
msno.bar(data)
```

```python
# Code snippet 8
# Dropping 'Home Prices Ranking', 'Community Health Ranking' columns due to missing data (more than 20%)
data80 = data.drop(['Home Prices Ranking', 'Community Health Ranking'], axis = 1)

# Displaying data80.columns
data80.columns
```

```python
# Code snippet 9
# Dropping NOISE variables - 'School', 'School URL',  'District URL',  'City URL','Phone'
predictors = data80.drop(['School', 'School URL',  'District URL',  'City URL','Phone'], axis = 1)

# Displaying predictors.columns
predictors.columns
```

```python
# Code snippet 10
# Converting Zip to a string variable
predictors['Zip'] = predictors['Zip'].astype(str)
```

```python
# Code snippet 12
# Importing data_profiling function from the fg-data-profiling package
import data_profiling
```

```python
# Code snippet 13
# Generating detailed data profiling report using the data_profiling function
profile  = data_profiling.ProfileReport(predictors, title = 'TX Schools Data Profiling Report')

# Converting raw profile report into an interactive HTML report
profile.to_file('TX Schools Data Profiling Report.html')
```

```python
min_students = predictors['Number Students'].min()
max_students = predictors['Number Students'].max()

print(f"Minimum Number of Students: {min_students}")
print(f"Maximum Number of Students: {max_students}")
```

```python
districts_high_free_lunch = predictors[predictors['Percent Free/Disc Lunch'] > 25]['District'].unique()

print("Districts where 'Percent Free/Disc Lunch' exceeds 25%:")
for district in districts_high_free_lunch:
    print(district)
```

```python
num_districts = len(districts_high_free_lunch)
print(f"Number of districts where 'Percent Free/Disc Lunch' exceeds 25%: {num_districts}")
```

```python
median_free_lunch = predictors['Percent Free/Disc Lunch'].median()
print(f"The median percentage of students receiving free or discounted lunches is: {median_free_lunch}%")
```

```python
print('Percentage of schools by type:')
for col in ['Is Title I', 'Is Charter', 'Is Magnet', 'Is Virtual']:
    # Convert 'Yes'/'No' to 1/0 for calculation
    predictors[col] = predictors[col].map({'Yes': 1, 'No': 0})
    percentage = predictors[col].mean() * 100
    print(f'- {col}: {percentage:.2f}%')
```
