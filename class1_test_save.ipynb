```python
import pandas as pd

# Define the URL of the CSV file
csv_url = "https://raw.githubusercontent.com/nvamsimohan/DallasDSA/refs/heads/main/PCCI%20Cases%20Readmit30%20Data.csv"

# Load the data into a pandas DataFrame
df = pd.read_csv(csv_url)

# Display the first few rows of the DataFrame
display(df.head())
```

### Column Explanations:

*   **`Admit_Week`**: The day of the week the patient was admitted (e.g., Sunday, Monday).
*   **`Admit_Month`**: The month the patient was admitted (e.g., Jan, Feb).
*   **`Gender`**: The gender of the patient.
*   **`Marital_Status`**: The marital status of the patient (e.g., Married, Single).
*   **`Insurance_Provider`**: The patient's health insurance provider (e.g., Medicare, Commercial).
*   **`Tobacco_User`**: Indicates if the patient uses tobacco (e.g., Never, Quit, Current).
*   **`Depression`**: Indicates if the patient has a history of depression (Yes/No).
*   **`ICU`**: Indicates if the patient was admitted to the Intensive Care Unit (Yes/No).
*   **`Drug_Abuse`**: Indicates if the patient has a history of drug abuse (Yes/No).
*   **`Mood_Disorder`**: Indicates if the patient has a mood disorder (Yes/No).
*   **`Temperature`**: The patient's body temperature at admission.
*   **`Pat_Pain_Score`**: The patient's self-reported pain score (likely on a scale, e.g., 0-10).
*   **`ER_Visits`**: The number of emergency room visits the patient had prior to this admission.
*   **`IP_Visits`**: The number of inpatient (hospital) visits the patient had prior to this admission.
*   **`Chronic_Conditions`**: The number of chronic health conditions the patient has.
*   **`Glucose`**: The patient's glucose level.
*   **`Condition`**: The primary medical condition for which the patient was admitted (e.g., Pneumonia, Heart_Failure).
*   **`Care_Plan_Following_Discharge`**: The type of care plan recommended for the patient after discharge (e.g., Telehealth, Skilled Nursing Facility).
*   **`Cost_Of_Initial_Stay`**: The monetary cost associated with the initial hospital stay.
*   **`Readmit30`**: The target variable, indicating whether the patient was readmitted within 30 days (1 = Yes, 0 = No).
